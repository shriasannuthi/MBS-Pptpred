```python
from datetime import datetime

def calculate_month_difference(date1, date2):
    date1 = datetime.strptime(date1, "%m%Y")
    date2 = datetime.strptime(date2, "%m%Y")

    months = (date2.year - date1.year) * 12 + (date2.month - date1.month)
    
    return months
```

```python
RMFM=[]

for i in range(len(data['Loan Identifier'])):
    RMFM.append(data.iloc[i,7]-calculate_month_difference(str(data.iloc[i,8]),str(data.iloc[i,1])))
```
