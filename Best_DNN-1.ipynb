```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('Dataset.csv')
```

```python
ic=[]

for i in range(len(df['Loan Identifier'])):
    r=df.iloc[i,4]/1200
    p=df.iloc[i,-2]

    ic.append(p*r)   


df['Interest_Component']=ic
```

```python
import pandas as pd

# Assuming you have a dataset in a DataFrame called 'df' with a column 'Loan_ID' that represents the loan identifier

# Group the data by Loan_ID
grouped_data = df.groupby('Loan Identifier')

# Initialize an empty list to store the filtered data
filtered_data = []

# Iterate over each group
for _, group in grouped_data:
    # Exclude the first six entries for each loan
    filtered_group = group.iloc[6:]
    filtered_data.append(filtered_group)

# Concatenate the filtered data back into a single DataFrame
filtered_df = pd.concat(filtered_data)

# Print the filtered loan data
filtered_df


```

```python
# saving the dataframe
filtered_df.to_csv('FilteredDataset.csv', header=True, index=False)
```

```python
filtered_df=filtered_df[filtered_df['Original UPB']!=filtered_df['Current Actual UPB']]
```

```python
df.drop(['Loan Identifier','Monthly Reporting Period','First Time Home Buyer Indicator','Occupancy Status','Interest Only Loan Indicator','Modification Flag','Zero Balance Code','Special Eligibility Program','Property Valuation Method','High Balance Loan Indicator','EMI','UPB_diff','Interest_Component','PREPAYMENT_AMT','Remaining Months To Maturity'], axis=1, inplace = True)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X=df.drop(['Current Actual UPB'],axis=1)
y = df['Current Actual UPB']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train.info()
```

```python
X_train
```

```python
y_train
```

```python
model = Sequential()

model.add(Dense(units=64, activation='relu', input_dim=X.shape[1]))

model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=16, activation='relu'))

model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mean_absolute_error')
```

```python
history = model.fit(X_train_scaled, y_train, epochs=50, batch_size=32) 
```

```python
predictions = model.predict(X_test_scaled)
```

```python
print(predictions)
```

```python
print(y_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(predictions, y_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(predictions, y_test))
```
