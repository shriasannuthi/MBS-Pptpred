```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('Dataset.csv')
df
```

```python
df.drop(['EMI','UPB_diff','Interest_Component','PREPAYMENT_AMT','Remaining Months To Maturity', 'Loan Identifier'], axis=1, inplace = True)
```

```python
y= df['Current Actual UPB']
```

```python
df.drop(['Current Actual UPB'], axis=1, inplace = True)
```

```python
df['Current Actual UPB'] = y
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X = df.iloc[:, :30]  
y = df.iloc[:, 30]  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
y_train
```

```python
model = Sequential()

model.add(Dense(units=64, activation='relu', input_dim=X.shape[1]))

model.add(Dense(units=16, activation='relu'))

model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mean_absolute_error' )
```

```python
model.fit(X_train, y_train, epochs=2, batch_size=32) 
```

```python
predictions = model.predict(X_test)
predictions = np.clip(predictions, 0, None)
```

```python
print(predictions)
```

```python
print(y_test)
```

```python
y_test= y_test.values
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(predictions, y_test)
```
